## Gamma

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Gamma

plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
#plt.subplot(1, 2, 2)

distribution = np.random.gamma(3,1, 100)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)


# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)
plt.title("n=100")

plt.subplot(1, 2, 2)

distribution = np.random.gamma(3,1, 10000)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)

# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)
plt.title("n=10,000")


plt.show()
```

Generally, the gamma distribution has a long tail on the right side, and thus the descriptive statistics will also be slightly skewed in this direction from the center of the peak in the data. 

## Binomial



```python
# Binomial Distribution

plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
#plt.subplot(1, 2, 2)

distribution = np.random.binomial(50,0.30, 100)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)
plt.title("n=100")
# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)

plt.subplot(1, 2, 2)

distribution = np.random.binomial(50,0.30, 10000)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)
plt.title("n=10,000")
# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)


plt.show()
```

The binomial distribution has more variability than some of the other distributions. The mean and standard deviation of the dataset provide a useful insight into the central tendencies of the data. The binomial distribution is not expected to be strongly skewed in one direction or another. 

## Poisson

```python
# Poisson Distribution

plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
#plt.subplot(1, 2, 2)

distribution = np.random.poisson(5, 100)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)
plt.title("n=100")
# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)


plt.subplot(1, 2, 2)

distribution = np.random.poisson(5, 10000)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)
plt.title("n=10,000")
# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)


plt.show()
```

Poisson distributions represent data where an event occurs between 0 and (technically) an infinite number of times. The distribution has a long tail to the right, and the measures of central tendency will thus also be shifted to the right.

## Uniform

```python
# Uniform Distribution

plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
#plt.subplot(1, 2, 2)

distribution = np.random.uniform(5,10, 100)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)
plt.title("n=100")
# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)


plt.subplot(1, 2, 2)

distribution = np.random.uniform(5,10, 100)
dist_mean = np.mean(distribution)
dist_std = np.std(distribution)

plt.hist(distribution)
plt.title("n=10,000")
# Show the mean
plt.axvline(dist_mean, color='c', linestyle='solid', linewidth=4)

# Show one standard deviation from the mean
plt.axvline(dist_mean + dist_std, color='c', linestyle='dashed', linewidth=2)
plt.axvline(dist_mean - dist_std, color='c', linestyle='dashed', linewidth=2)

plt.show()
```

The data within a uniform distribution has an equal probability of occurring anywhere within the low and high values. On average, the mean value will occur exactly between the high and low value, and the calculated mean and standard deviation from the dataset will elucidate whether the dataset is equally distributed.

## Geometric
